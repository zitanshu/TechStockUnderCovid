```python
# Grangerâ€™s Causality Test
from statsmodels.tsa.stattools import grangercausalitytests
maxlag=12
test = 'ssr_chi2test'
def grangers_causation_matrix(data, variables, test='ssr_chi2test', verbose=False):    
    """Check Granger Causality of all possible combinations of the Time series.
    The rows are the response variable, columns are predictors. The values in the table 
    are the P-Values. P-Values lesser than the significance level (0.05), implies 
    the Null Hypothesis that the coefficients of the corresponding past values is 
    zero, that is, the X does not cause Y can be rejected.

    data      : pandas dataframe containing the time series variables
    variables : list containing names of the time series variables.
    """
    df = pd.DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in df.columns:
        for r in df.index:
            test_result = grangercausalitytests(data[[r, c]], maxlag=maxlag, verbose=False)
            p_values = [round(test_result[i+1][0][test][1],4) for i in range(maxlag)]
            if verbose: print(f'Y = {r}, X = {c}, P Values = {p_values}')
            min_p_value = np.min(p_values)
            df.loc[r, c] = min_p_value
    df.columns = [var + '_x' for var in variables]
    df.index = [var + '_y' for var in variables]
    return df

print("Grangerâ€™s Causality Test for Stock Data Before the Pandemic:")
grangers_causation_matrix(all_stock_before, variables = all_stock_before.columns) 
```

```python
# Cointegration Test
from statsmodels.tsa.vector_ar.vecm import coint_johansen

def cointegration_test(df, alpha=0.1): 
    """Perform Johanson's Cointegration Test and Report Summary"""
    out = coint_johansen(df,-1,5)
    d = {'0.9':0, '0.95':1, '0.99':2}
    traces = out.lr1
    cvts = out.cvt[:, d[str(1-alpha)]]
#     print(out.cvt)
    def adjust(val, length= 6): return str(val).ljust(length)

    # Summary
    print('Name   ::  Test Stat > C(90%)    =>   Signif  \n', '--'*20)
    for col, trace, cvt in zip(df.columns, traces, cvts):
        print(adjust(col), ':: ', adjust(round(trace,2), 9), ">", adjust(cvt, 8), ' =>  ' , trace > cvt)

print("Cointegration Test for Stock Data Before the Pandemic:")
cointegration_test(all_stock_before)
```

```python
#train test split
num_obs = 10 # predict for next 10 days
all_stock_before_train, all_stock_before_test = all_stock_before[0:-num_obs], all_stock_before[-num_obs:]

# Check size
print(all_stock_before_train.shape) 
print(all_stock_before_test.shape)
```

```python
# ADF Test on each column to check sataionalities - round 1
for name, column in all_stock_before_train.iteritems():
    DF_test(column, name=column.name)
    print('\n')
```

```python
#all variable are non-stationary, take 1st difference, then re-run ADF test
all_stock_before_differenced = all_stock_before_train.diff().dropna()
# all_stock_before_differenced = all_stock_before_differenced.diff().dropna()

for name, column in all_stock_before_differenced.iteritems():
    DF_test(column, name=column.name)
    print('\n')
```

All p-value < 0.05 after first difference, ready to move on

```python
#VAR model lag selection and training
model_before = VAR(all_stock_before_differenced)
AIC = []
BIC =  []
FPE = []
HQIC = []
for i in [1,2,3,4,5,6,7,8,9]:
    result = model_before.fit(i)
    AIC.append(result.aic)
    BIC.append(result.bic)
    FPE.append(result.fpe)
    HQIC.append(result.hqic)
#     print('Lag Order =', i)
#     print('AIC : ', result.aic)
#     print('BIC : ', result.bic)
#     print('FPE : ', result.fpe)
#     print('HQIC: ', result.hqic, '\n')
    
x_value = [1,2,3,4,5,6,7,8,9]
plt.plot(x_value, AIC, color = 'black', label = 'aic')
plt.plot(x_value, BIC, color = 'red', label = 'bic')
# plt.plot(FPE, color = 'green', label = 'fpe')
plt.plot(x_value, HQIC, color = 'blue', label = 'hqic')
plt.legend(loc="upper left")
plt.title("Lag Selection for Model with Stock Data Before the Pandemic")
# print("Lag Selection for Model with Stock Data Before the Pandemic: ")
```

```python
x = model_before.select_order(maxlags=12)
x.summary()
```

```python
#use lag 1 as selected model to fit data
model_before_fitted = model_before.fit(1)
model_before_fitted.summary()
```

```python
#check for residuals
from statsmodels.stats.stattools import durbin_watson
out = durbin_watson(model_before_fitted.resid)

for col, val in zip(all_stock_before.columns, out):
    print(col, ':', round(val, 2))
```

```python
# Forecasting
# Get the lag order
lag_order_before = model_before_fitted.k_ar
# print(lag_order_before) 

# Input data for forecasting
forecast_before_input = all_stock_before_differenced.values[-lag_order_before:]
forecast_before_input

# Forecast - this is the result in terms of 1st differencing
fc_before = model_before_fitted.forecast(y=forecast_before_input, steps=num_obs)
all_stock_before_forecast = pd.DataFrame(fc_before, index=all_stock_before.index[-num_obs:], columns=all_stock_before.columns + '_1diff')
all_stock_before_forecast
```

```python
#transformation invertion back to original scale
def invert_transformation(df_train, df_forecast, second_diff=False):
    """Revert back the differencing to get the forecast to original scale."""
    df_fc = df_forecast.copy()
    columns = df_train.columns
    for col in columns:        
        # Roll back 1st Diff
        df_fc[str(col)+'_forecast'] = df_train[col].iloc[-1] + df_fc[str(col)+'_1diff'].cumsum()
    return df_fc
all_stock_before_results = invert_transformation(all_stock_before_train, all_stock_before_forecast)        
all_stock_before_results
```

```python
#plot for forecast result and actual result
print("Forecast Results vs. Actuals with Stock Data Before the Pandemic:")
fig, axes = plt.subplots(nrows=int(len(all_stock_before.columns)/2), ncols=2, dpi=150, figsize=(10,10))
for i, (col,ax) in enumerate(zip(all_stock_before.columns, axes.flatten())):
    all_stock_before_results[col+'_forecast'].plot(legend=True, ax=ax).autoscale(axis='x',tight=True)
    all_stock_before_test[col][-num_obs:].plot(legend=True, ax=ax);
    ax.set_title(col + ": Forecast vs Actuals")
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines["top"].set_alpha(0)
    ax.tick_params(labelsize=6)

plt.tight_layout();
```

```python
# evaluation
from statsmodels.tsa.stattools import acf
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None], 
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    return({'mape':mape, 'me':me, 'mae': mae, 
            'mpe': mpe, 'rmse':rmse, 'corr':corr, 'minmax':minmax})

res_before = []
name_list = list(all_stock_before.columns)
for name in name_list:
#     print('Forecast Accuracy of: {}'.format(name))
    accuracy_prod = forecast_accuracy(all_stock_before_results[name+'_forecast'].values, all_stock_before_test[name])
#     print(accuracy_prod)
    res_before.append(accuracy_prod)
#     for k, v in accuracy_prod.items():
#         print(k, ': ', round(v,4))
#     print('\n')
print("Evaluation of Model Trained with Stock Data Before the Pandemic:")
res_before = pd.DataFrame.from_dict(res_before, orient='columns')
res_before['Company'] = list(all_stock_before.columns)
res_before = res_before.set_index('Company')
res_before
```

```python
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, GRU, LSTM, Bidirectional,Dropout
from keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import SGD
```

```python
num_obs = 20

def MinMaxScale(data,time_steps,for_periods,num_obs):
    '''normalize the data before feeding to RNN/BiLSTM'''
    
    
    # predict for next 20 days, as test
    train, test = data[0:-num_obs].iloc[:,0:1].values, data[-num_obs:].iloc[:,0:1].values
    train_len = len(train)
    test_len = len(test)
#     print(train_len, test_len) # 242 10

    # scale the data
    sc = MinMaxScaler(feature_range=(0,1))
    data_scaled = sc.fit_transform(train)

    # create training data of s samples and t time steps
    X_train = []
    y_train = []
    y_train_stacked = []
    for i in range(time_steps,train_len-1): 
        X_train.append(data_scaled[i-time_steps:i,0])
        y_train.append(data_scaled[i:i+for_periods,0])
    X_train, y_train = np.array(X_train), np.array(y_train)
#     print(y_train)

    # Preparing to create X_test
    X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
    
    inputs = pd.concat((data[0:-num_obs],data[-num_obs:]),axis=0).values
    inputs = inputs[len(inputs)-len(test) - time_steps:]
    inputs = inputs.reshape(-1,1)
    inputs  = sc.transform(inputs)

    # Preparing X_test
    X_test = []
    for i in range(time_steps,test_len+time_steps-for_periods):
        X_test.append(inputs[i-time_steps:i,0])
        
    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

    return X_train, y_train , X_test, sc

def rnn_model_function(X_train, y_train, X_test, sc):
    '''
    single layer rnn
    '''
    # create a model
    
    rnn_model = Sequential()
    rnn_model.add(SimpleRNN(32, return_sequences=True))
    rnn_model.add(SimpleRNN(32, return_sequences=True))
#     rnn_model.add(SimpleRNN(32, return_sequences=True))
    rnn_model.add(SimpleRNN(32))
    rnn_model.add(Dense(2)) # The time step of the output: 2

    rnn_model.compile(optimizer='rmsprop', loss='mean_squared_error')

    # fit the RNN model
    rnn_model.fit(X_train, y_train, epochs=100, batch_size=150, verbose=0)

    # Finalizing predictions
    rnn_predictions = rnn_model.predict(X_test)
    rnn_predictions = sc.inverse_transform(rnn_predictions) #reverse the min max scaler

    return rnn_model, rnn_predictions

def actual_pred_plot(data, preds,to_plot):
    '''Plot the actual vs. prediction'''
    actual_pred = pd.DataFrame(columns = ['Adj. Close', 'prediction'])
    actual_pred['Adj. Close'] = data.iloc[-len(preds):,0].to_list()
    actual_pred['prediction'] = preds[:,0]

    m = RootMeanSquaredError()
    m.update_state(np.array(actual_pred['Adj. Close']),np.array(actual_pred['prediction']))
    if to_plot == True:
        return (m.result().numpy(), actual_pred.plot() )
    else:
        return m.result().numpy()
```

```python
#AAPL before
AAPL_before = pd.DataFrame().assign(AAPL=all_stock_before['AAPL'])
X_train, y_train, X_test, sc = MinMaxScale(AAPL_before,5,2,num_obs)
best_pred = []
min_RMSE = float('inf')
for i in range(10):
    rnn_model, rnn_predictions = rnn_model_function(X_train, y_train, X_test, sc)
    rmse = actual_pred_plot(AAPL_before,rnn_predictions,False)
    if rmse < min_RMSE:
        min_RMSE = rmse
        best_pred = rnn_predictions
    
actual_pred_plot(AAPL_before,best_pred,True)
```

```python
def LSTM_model(X_train, y_train, X_test, sc):
    
    # The LSTM architecture
    my_LSTM_model = Sequential()
    my_LSTM_model.add(Bidirectional(LSTM(units=128, return_sequences=True, input_shape=(X_train.shape[1],1), activation='tanh')))
    my_LSTM_model.add(Bidirectional(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1],1), activation='tanh')))

#     my_LSTM_model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1],1), activation='tanh'))
    #my_LSTM_model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1],1), activation='tanh'))
    my_LSTM_model.add(Bidirectional(LSTM(units=64, activation='tanh')))
    my_LSTM_model.add(Dropout(0.2))
    my_LSTM_model.add(Dense(units=2))

    # Compiling
    my_LSTM_model.compile(optimizer='adam',loss='mean_squared_error')
    # Fitting to the training set
    my_LSTM_model.fit(X_train,y_train,epochs=50,batch_size=128, verbose=0)

    LSTM_prediction = my_LSTM_model.predict(X_test)
    LSTM_prediction = sc.inverse_transform(LSTM_prediction)

    return my_LSTM_model, LSTM_prediction
```
